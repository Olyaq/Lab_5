```python
import numpy as np  # библиотека дляработы с массивами и мат.операциями
import matplotlib.pyplot as plt  # бмблиотека для визуализации

# Генерирует данные
x = np.linspace(0, 10, 100) # создает массив х из 100т. от 0 до 10
y = np.sin(x) # вычисляет синус от х
z = np.cos(x) #вычисляет косинус от х

# Строит графики
plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')

# Добавление заголовка
plt.title('Графики sin(x) и cos(x)') # настраивает график
plt.xlabel('x') # добавил заголовок
plt.ylabel('y')

# Легенда и сетка
plt.legend() #отображает легенду с метками синуса и косинуса
plt.grid() # включает сетку

# Показать график
plt.show()
```

```python
# Импорт необходимых библиотек
import numpy as np  # Библиотека для мат.операций и работы с массивами
import matplotlib.pyplot as plt  # Библиотека для построения графиков

# Генерация случайных данных
data = np.random.normal(size=1000) # создаем массив из 1000 значений, среднее 0, стандартное откл. 1

# Построение гистограммы
plt.hist(data, bins=20) # фун.постороение, массив, кол-во столбцов

# Настройки графика
plt.title('Гистограмма нормального распределения') # заголовок
plt.xlabel('Значения') # подпись
plt.ylabel('Частота')

plt.grid(True) # включение сетки
plt.show() # вывод Гистограммы
```

```python
import numpy as np  #Библиотека для математических операций и работы с массивами

# Создаем матрицу 5x5 со случайными числами от 1 до 10
matrix = np.random.randint(1, 11, (5, 5)) # мин.1, макс.11 (не включается поэтому до 10), размер матрицы

# Вычисляем характеристики
print("Матрица:\n", matrix) # вывод матрицы, перенос строки перед выводом матрицы
print("\nСреднее:", np.mean(matrix)) # вычисление и вывод сред.знач, вычисляет среднее всех ариф.матрицы
print("Максимум:", np.max(matrix)) # находит и выводит максимальное, возвращает максимальное знач
print("Минимум:", np.min(matrix)) # вычисляет и вывод минимальное
print("Сумма столбцов:", np.sum(matrix, 0)) #вычисляет и вывод из столбцов, суммирует,сум.вдоль столбцов
```

```python
import numpy as np # импорт библиотеки для работы с массивами и мат.операциями
import matplotlib.pyplot as plt # импорт модуля из библиотеки

# Создаем матрицу 5x5
matrix = np.random.randint(1, 11, (5, 5))

# Выводим матрицу
print("Матрица:\n", matrix)

# Строим тепловую карту
plt.imshow(matrix, cmap='viridis') #создаем, отображает матрицу как изображение, данные для показа, использование цветов
plt.colorbar() # создает шкалу использования цветов знач.матрицы

# Добавляем числа в ячейки
for i in range(5): # индекс строки
    for j in range(5): # индекс
        plt.text(j, i, matrix[i,j], #добавляем текст в ячейку, отображает значение
                 ha='center', va='center', # горизонтальное и вертикальное по центру
                 color='w' if matrix[i,j] > 5 else 'k') # цвет текста зависит от знач.

plt.show() # показывает график
```

```python
import numpy as np # Библиотека для математических вычислений
import matplotlib.pyplot as plt # Библиотека для построения графиков

# Создаем 3 графика в одном ряду
fig, ax = plt.subplots(1, 3, figsize=(12, 3)) # объект всего изображения, массив из 3 осей, создаем область, размер всего изображения

# 1. График линейной функции
x = np.linspace(0, 5, 30) # создаем массив x с 30 точками от 0 до 5
ax[0].plot(x, x**2, 'r') # строим график y = x^2, красная линия
ax[0].set_title('y = x²') # устанавливаем заголовок график

# 2. Точечный график
x, y = np.random.rand(2, 20) # генерируем 2 массива по 20 случайных чисел от 0 до 1
ax[1].scatter(x, y, c='g') # рисуем точки, зеленый цвет
ax[1].set_title('Случайные точки') # устанавливаем заголовок

# 3. Столбчатая диаграмма
ax[2].bar(['A','B','C'], [3,7,2]) # рисуем столбцы с высотами
ax[2].set_title('Категории') # устанавливаем заголовок

plt.tight_layout() # делаем отступы между графиками
plt.show() #выводим графики
```
