```python
import numpy as np  #      .
import matplotlib.pyplot as plt  #   

#  
x = np.linspace(0, 10, 100) #     100.  0  10
y = np.sin(x) #    
z = np.cos(x) #   

#  
plt.plot(x, y, label='sin(x)')
plt.plot(x, z, label='cos(x)')

#  
plt.title(' sin(x)  cos(x)') #  
plt.xlabel('x') #  
plt.ylabel('y')

#   
plt.legend() #      
plt.grid() #  

#  
plt.show()
```

```python
import numpy as np  #       

#   5x5     1  10
matrix = np.random.randint(1, 11, (5, 5)) # .1, .11 (    10),  

#  
print(":\n", matrix) #  ,     
print("\n:", np.mean(matrix)) #    .,    .
print(":", np.max(matrix)) #    ,   
print(":", np.min(matrix)) #    
print(" :", np.sum(matrix, 0)) #    , ,. 
```

```python
import numpy as np #        .
import matplotlib.pyplot as plt #    

#   5x5
matrix = np.random.randint(1, 11, (5, 5))

#  
print(":\n", matrix)

#   
plt.imshow(matrix, cmap='viridis') #,    ,   ,  
plt.colorbar() #     .

#    
for i in range(5): #  
    for j in range(5): # 
        plt.text(j, i, matrix[i,j], #   ,  
                 ha='center', va='center', #     
                 color='w' if matrix[i,j] > 5 else 'k') #     .

plt.show() #  
```

```python
import numpy as np #    
import matplotlib.pyplot as plt #    

#  3    
fig, ax = plt.subplots(1, 3, figsize=(12, 3)) #   ,   3 ,  ,   

# 1.   
x = np.linspace(0, 5, 30) #   x  30   0  5
ax[0].plot(x, x**2, 'r') #   y = x^2,  
ax[0].set_title('y = x') #   

# 2.  
x, y = np.random.rand(2, 20) #  2   20    0  1
ax[1].scatter(x, y, c='g') #  ,  
ax[1].set_title(' ') #  

# 3.  
ax[2].bar(['A','B','C'], [3,7,2]) #    
ax[2].set_title('') #  

plt.tight_layout() #    
plt.show() # 
```
